```python
import pandas as pd

# Read CSV data into dataframe. Pandas automatically uses the first row as column names
df = pd.read_csv('./files/sales_records.csv')  

# Get statistics on the numeric columns (use `df.describe(include='O')` for text columns)
print(df.describe())  
print()

# Get information on all the columns ('object' means text/string)
print(df.info())  
print()

# Display first 5 rows of the dataframe (`df.describe(__n__)` displays n rows)
print(df.head(5))  

df['total_sales'] = df['Units Sold'] * df['Unit Price']
print(df)

print(df.info())
print(df.describe())
```
